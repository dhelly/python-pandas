## Export `DataFrame` to CSV File with the `.to_csv()` Method

```python
baby_names = pd.read_csv("https://data.cityofnewyork.us/resource/25th-nujf.csv?$query=SELECT%20%60brth_yr%60%2C%20%60gndr%60%2C%20%60ethcty%60%2C%20%60nm%60%2C%20%60cnt%60%2C%20%60rnk%60%0AORDER%20BY%0A%20%20%60brth_yr%60%20DESC%20NULL%20FIRST%2C%0A%20%20%60gndr%60%20ASC%20NULL%20LAST%2C%0A%20%20%60ethcty%60%20ASC%20NULL%20LAST%2C%0A%20%20%60rnk%60%20ASC%20NULL%20LAST")
baby_names.head(3)
```

```python
#cuidado para nÃ£o sobre escrever o arquivo original
baby_names.to_csv("Baby Names.csv", index = False, columns = ["gndr","ethcty", "nm"], encoding = "utf-8")
```

## Export Excel File

```python
baby_names = pd.read_csv("https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD")
baby_names.head(3)
```

```python
girls = baby_names[baby_names["Gender"] == "FEMALE"]
boys = baby_names[baby_names["Gender"] == "MALE"]
```

```python
girls
```

```python
boys
```

```python
excel_file = pd.ExcelWriter("Baby_names.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name="Grils", index = False)
```

```python
boys.to_excel(excel_file, sheet_name="Boys", index = False, columns=["Year of Birth", "Gender", "Ethnicity"])
```

```python
excel_file.close()
```

## Export Excel File

```python
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names = pd.read_csv(URL)
baby_names.head(3)
```

```python
girls = baby_names[baby_names["GNDR"] == "FEMALE"]
boys = baby_names[baby_names["GNDR"] == "MALE"]
```

```python
excel_file = pd.ExcelWriter("Baby Names.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name = "Girls", index = False)
boys.to_excel(excel_file, sheet_name = "Boys", index = False, columns = ["GNDR", "NM", "CNT"])
```

```python
excel_file.save()
```
